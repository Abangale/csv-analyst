```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
'''(1A) Jumlah Kasus Setiap Negara'''
cases = pd.read_csv('Dengue_Daily_EN.csv', sep = ',', parse_dates = [0])
cases = cases.groupby('County_living').sum()
cases = cases.sort_values(by='Number_of_confirmed_cases', ascending=False)
plt.figure()
cases.plot.barh()
plt.title('Total Dengue Cases in Chinese')
```

```python
'''(1B) Jumlah Kasus Setiap Bulan'''
month = pd.read_csv('Dengue_Daily_EN.csv', sep = ',', parse_dates = [0])
month['MonthOfTime'] = month['Date_Onset'].dt.month_name()
month = month.groupby('MonthOfTime').sum()
month = month.sort_values(by='Number_of_confirmed_cases', ascending=False)
plt.figure()
month.plot.barh()
plt.title('Total Dengue Cases Within Months')
```

```python
'''(1C) Jumlah Kasus Setiap Grup Ruangan'''
ages = pd.read_csv('Dengue_Daily_EN.csv', sep = ',', parse_dates = [0])
ages = ages.groupby('Age_Group').sum()
ages = ages.sort_values(by='Number_of_confirmed_cases', ascending=False)
plt.figure()
ages.plot.barh()
plt.title('Total Cases by Age Group')
```

```python
'''(2) Plot Jumlah Kasus per Bulan dan Bulan apa paling banyak terjadi kasus'''
max_month = pd.read_csv('Dengue_Daily_EN.csv', sep = ',', parse_dates = [0])
max_month['MonthOfTime'] = max_month['Date_Onset'].dt.month_name()
max_month = max_month.groupby('MonthOfTime').agg({'Number_of_confirmed_cases': 'sum'}).reset_index().sort_values(by='MonthOfTime', ascending=True)
def highlight_max(n):
    max_line = n.max()
    return ['background-color:blue' if v == max_line else '' for v in n]

max_month.style.apply(highlight_max, subset = ['MonthOfTime','Number_of_confirmed_cases'], axis=0)
```

```python
'''(3) Menghitung yang menunjukkan jumlah kasus setiap tahun'''
year = pd.read_csv('Dengue_Daily_EN.csv', sep = ',', parse_dates = [0])
year['YearOfTime'] = year['Date_Onset'].dt.year
year = year.groupby('YearOfTime').sum()
year
```
